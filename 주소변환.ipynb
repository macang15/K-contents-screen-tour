```python
import time, random, re
import requests
from bs4 import BeautifulSoup
import openpyxl
import pandas as pd
import numpy as np
```

```python
#도로명주소 안내시스템 사이트를 통한 도로명주소 -> 지번주소 변환

def doro_to_jibun(keyword):
    url = f"https://www.juso.go.kr/support/AddressMainSearch.do?searchKeyword={keyword}&dsgubuntext=&dscity1text=&dscounty1text=&dsemd1text=&dsri1text=&dssan1text=&dsrd_nm1text="
    req = requests.get(url)
    soup = BeautifulSoup(req.text, "html.parser")
    soup1 = soup.select('#searchAddress > div.container.support_search_list > div.search_list > ol > li')
    
    if len(soup1)==0:  # 도로명주소를 검색했을 때 결과가 나오지 않는다면
        return np.nan
    else:
        jibun_juso = soup1[0].find('div', class_='subejct_2').find('span', class_='roadNameText').text
        jibun = re.sub('&nbsp;|\t|\r|\n', '', jibun_juso).strip().replace('  ', ' ')

        time.sleep(0.5)
        return jibun
```
